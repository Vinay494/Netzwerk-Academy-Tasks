```python
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
x=(jobs['Fiscal Year'])
y=(jobs['Public Sector Employees (million)'])
z=(jobs['Private Sector Employees (million)'])
p=(jobs['No. of Emp. on Live Reg. (million)'])
q=(jobs['Population (million)'])
r=(jobs['Prompts_Time'])

plt.subplot(1,2,1) # To create a first sub plot
plt.scatter(q,y,c = 'g',marker = '.') #to create a scatter plot
plt.xlabel('Population') #xlabel
plt.ylabel('Public Sector Employees') #ylabel


plt.subplot(1,2,2) # To create a first sub plot
plt.scatter(q,z,c = 'r',marker = '*') #to create a scatter plot
plt.xlabel('Population') #xlabel
plt.ylabel('Private Sector Employees') #ylabel
plt.title('Public vs Private employees count rise with raise in population') #title of the graph
plt.show() #to see the plot
```

# Interference from above plot

1) As the population increases till some where around 900 million, the public sector employees count was incresing, but there after it got reduced a bit.

2) But private sector employees count was fluctuating, but at the end finally it went to peak but the growth and fall was zigzag.
